```python
def age_groups(row):
    if row["age"] < 30:
        return "20 - 29"
    elif row["age"] < 40:
        return "30 - 39"
    elif row["age"] < 50:
        return "40 - 49"
    elif row["age"] < 60:
        return "50 - 59"
    else:
        return "60 - 70"
```

```python
data["age range"] = data.apply(age_groups, axis=1)
```

```python
X = data[['age', 'annual Salary', 'net worth']]
y = data[['car purchase amount']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()

X_train  = scaler.fit_transform(X_train)
```

```python
X_train 
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
X_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
def ModelResult(prediction):
    print("Mean absolute error on the model is {}:".format(mean_absolute_error(y_test, prediction)))
    print("Mean absolute error on the model is {}:".format(np.sqrt(mean_squared_error(y_test, prediction))))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
prediction_linear_regression = lr.predict(X_test)
```

```python
ModelResult(prediction_linear_regression)
```

```python
from sklearn.svm import SVR
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[0.01, 0.1, 0.5], "kernel": ["linear", "rbf", "poly"], "degree": [2,3,4]}
```

```python
from sklearn.model_selection import GridSearchCV

gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, y_train.values.ravel())
```

```python
prediction_gridsvr = gridsvr.predict(X_test)
```

```python
ModelResult(prediction_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor

rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth": [5,10,15], "n_estimators": [2,3,4,8,9,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
prediction_rfr = gridrfr.predict(X_test)
```
